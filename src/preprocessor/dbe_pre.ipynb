```python
import json

fold1_train_path = "/workspace/Deep_knowledge_tracing_baseline/datasets/dbe_kt22/fold1_train.json"
fold1_test_path = "/workspace/Deep_knowledge_tracing_baseline/datasets/dbe_kt22/fold1_test.json"

with open(fold1_train_path) as f:
    train_data = json.load(f)

with open(fold1_test_path) as f:
    test_data = json.load(f)
```

```python
train_data[0].keys()
```

```python
import pandas as pd

train_user_ids = []
train_skill_ids = []
train_corrects = []
train_item_ids = []

for data in train_data:
    for i in range(len(data["concept_ids"])):
        train_user_ids.append(data["user_id"])
        train_skill_ids.append(data["concept_ids"][i])
        train_corrects.append(data["response_ids"][i])
        train_item_ids.append(data["question_ids"][i])

train_df = pd.DataFrame({
    "user_id": train_user_ids, 
    "skill_id": train_skill_ids, 
    "correct": train_corrects, 
    "item_id": train_item_ids
})

train_df.head()
```

```python
# test data

test_user_ids = []
test_skill_ids = []
test_corrects = []
test_item_ids = []

for data in test_data:
    for i in range(len(data["concept_ids"])):
        test_user_ids.append(data["user_id"])
        test_skill_ids.append(data["concept_ids"][i])
        test_corrects.append(data["response_ids"][i])
        test_item_ids.append(data["question_ids"][i])

test_df = pd.DataFrame({
    "user_id": test_user_ids, 
    "skill_id": test_skill_ids, 
    "correct": test_corrects, 
    "item_id": test_item_ids
})

test_df.head()
```

```python
# concat
total_df = pd.concat([train_df, test_df], axis=0)

total_df.head()
```

```python
# save

total_df.to_csv("/workspace/Deep_knowledge_tracing_baseline/datasets/dbe_kt22/total.tsv", index=False, sep="\t")
```
